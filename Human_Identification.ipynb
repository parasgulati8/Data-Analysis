```
results_df = pd.DataFrame(acc, index=list(range(1,41)), columns=['Val Accuracy'])
```

```
ax = results_df.plot(kind='bar', legend=None, figsize=(25, 10), rot=0, fontsize=16)
for p in ax.patches:
    ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x() * 0.99, p.get_height() * 1.02), size=12)
```

Movements.jpg

```
mean = acc.mean()
std = acc.std()

gestures = []
for i in range(acc.shape[0]):
  if acc[i] >= mean+std:
    gestures.append(i)
```

```
gestures
```

```
ax = results_df.iloc[gestures, :].plot(kind='bar', legend=None, figsize=(25, 10), rot=45, fontsize=16)
for p in ax.patches:
    ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x() * 0.99, p.get_height() * 1.02), size=12)
```
