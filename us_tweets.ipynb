```
# Loading the excel file with all the US locations (manually filtered)
df = pd.read_excel('Copy of Locations.xlsx', header=None)
us_loc = (df[0]+df[1]).values
np.savetxt("us_loc.txt", us_loc, fmt='%s')
```

```
filename = 'us_loc.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
# split into words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
```

```
#remove special charaters
words = [word.lower() for word in tokens if word.isalpha()]
```

```
#remove duplicates
words = list(set(words))
```

```
# additional locations that were not present 
words.append('united states')
words.append('brooklyn')
words.append('los angeles')
words.append('san francisco')

words.remove('new')
words.remove('york')
words.append('new york')

words = list(set(words))
```

```
to_drop = []

def check(loc):
  if loc.lower() in words:
    return True
  # splitting loc and checking all the words in loc
  for element in re.findall(r'[a-zA-z]+', loc.lower()):
    if element in words:
      return True
  return False


for i in range(len(data)):
  if (check(data.iloc[i,4]) ):
    continue
  else:
    print('dropping', data.iloc[i,4], i)
    to_drop.append(i)
```

```
data.drop(index=to_drop, inplace=True)
```

```
data.to_csv('US Tweets.csv', encoding='utf-8')
```

# Key Words Extraction

```
def read_data():
    df = pd.read_csv("US Tweets.csv")
    return df

def write_output(df):
    df.to_csv("US Tweets with keyword tagged.csv")
```

```
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
from spacy.matcher import PhraseMatcher

def keyWordMatching(tweets_df,category, filename):
    keyword_dict = pd.read_csv(filename)
    CompDiscr_words = [nlp(text) for text in keyword_dict['cure'].dropna(axis = 0)]
    matcher = PhraseMatcher(nlp.vocab)
    matcher.add(category, None, *CompDiscr_words)
    customcolname = category+"_keywords"
    
    for ind in tweets_df.index:
        #text = tweets_df['tweet'][ind].replace("\\n", "")
        text = tweets_df['tweet'][ind].lower()    
        doc = nlp(text)

        d = []  
        matches = matcher(doc)

        for match_id, start, end in matches:
            rule_id = nlp.vocab.strings[match_id]  # get the unicode ID, i.e. 'COLOR'
            span = doc[start : end]  # get the matched slice of the doc
            d.append((span.text))  
        tweets_df[customcolname][ind] = d
        if(ind!=0 and ind % 10000==0):
            print("#### Completed validation of "+str(ind)+" records")
            
    return (tweets_df)



#x=keyWordMatching("Trump administration cuts funding for coronavirus researcher, jeopardizing possible COVID-19 cure - 60 MInutes - CBS News https://t.co/ZMZuYVar8o","cure", "keywords.csv")
#print(x)
```

```
us_tweets = read_data()
us_tweets["cure_keywords"] = ""
us_tweets["nature_of_virus_keywords"] = ""
us_tweets["conspiracy_theories_keywords"] = ""
us_tweets["emergency_measures_keywords"] = ""
us_tweets["good_news_keywords"] = ""
us_tweets["misbehaviour_scams_keywords"] = ""
write_output(keyWordMatching(us_tweets, "cure", "cure_keywords.csv"))
write_output(keyWordMatching(us_tweets, "nature_of_virus", "nature_keywords.csv"))
write_output(keyWordMatching(us_tweets, "conspiracy_theories", "conspiracy_keywords.csv"))
write_output(keyWordMatching(us_tweets, "emergency_measures", "emergency_keywords.csv"))
write_output(keyWordMatching(us_tweets, "good_news", "goodnews_keywords.csv"))
write_output(keyWordMatching(us_tweets, "misbehaviour_scams", "misbehaviour_keywords.csv"))
```
